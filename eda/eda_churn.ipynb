## Bivariate Analysis

### Hipóteses
- Clientes com Contrato tipo Mensal são mais propensos ao Churn

```python
# Check correlation between churn and contract tenure

# First, build a contingent table for contract tenure and churn
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name="Total")
```

```python
# Build a contingent table for contract tenure and churn with percents
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name="Total")
```

- Considerando o público total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham o contrato mensal. Desta forma podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
# Check the correlation of two qualitative dimensions

# Hypothesis test - Chi-square (Pearson)
# Here, we write two hypotheses:
# H0: Null hypothesis: both variables are independent
# H1: Complementar hypothesis: both variables are dependent
# This test try to confirm or refute the H0
# When the p-value is less than 0.05, we can reject the H0 and conclude that both variables are dependent
```

```python
# Create a crosstab DF
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Get scores and p-values
from scipy.stats import chi2_contingency
```

```python
# Tuple - immutable lists
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Present user friendly numbers
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
# Create a DF
df_chi_scores_churn_contract = pd.DataFrame({'Chi-Square': scores_churn_contract, 'p-values': pvalues_churn_contract})
df_chi_scores_churn_contract
```

- Conforme teste estatístico (Chi-Square), o p-value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Chi-Square alto, podemos afirmar que há uma forte correlação.

```python
# Avaliar a correlação entre uma variável qualitativa e quantitativa
# Usar o mesmo teste de Chi-Square, mas transformar a variável quantitativa em qualitativa.
# Hipótese: Cliente com menos de 6 meses de contrato é mais propenso ao Churn

# Criar uma nova coluna no DF com base em uma condição de outra coluna (TenureBelow6mo)
```

```python
# Check DF
df_churn.head(5)
```

- Como o p-value <=0.05, rejeitamos H0, ou seja, as variáveis não são independentes

```python
df_churn_month.TotalCharges.hist()
```

```python
# Turkey method - IQR (Non-normal distribution)
# IQR - Interquartile Range 
# Interquartile Range is the difference between the 75th and 25th data points
```

```python
# Define limits for the outliers
def outlier_limits(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.25)
    quartile3 = dataframe[variable].quantile(0.75)
    irq = (quartile3 - quartile1) * 1.5
    limits = (quartile1 - irq, quartile3 + irq)
    return limits

limits = outlier_limits(df_churn_month, 'TotalCharges')
limits
```

```python
df_churn_month[(df_churn_month.TotalCharges < limits[0]) | (df_churn_month.TotalCharges > limits[1])]
```
