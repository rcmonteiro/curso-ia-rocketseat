### Data transformation
> 5   MonthlyCharges    7043 non-null   float64
> 
> 6   TotalCharges      7043 non-null   object 

TotalCharges is on a wrong type, let's change to float64

```python
# Error, because some data can have whitespaces
# df_contract.TotalCharges = df_contract.TotalCharges.astype(float)
```

```python
# Transform the column from string to float, using errors=coerce
# coerce will change to null all values that cannot be converted to float
df_contract.TotalCharges = pd.to_numeric(df_contract.TotalCharges, errors='coerce')

# MonthlyCharges    7043 non-null   float64
# TotalCharges      7032 non-null   float64
# We lost 9 rows because they could not be converted to float! (9 missing values)
df_contract.info()
```

## Combining DataFrames for EDA

We have Customers, Contracts and Services in separated files, but for a better analysis it would be better if all the data were on a single DataFrame

```python
# Preparing DataFrame before combining them
df_services.rename(columns={'customerID': 'CustomerId'}, inplace=True)
```

```python
# Merge Customer and Services
df_temp = df_customer.merge(df_services, on='CustomerId')
df_temp.info()
```

```python
# Merge Temp and Contracts, we also can use left_on and right_on to specify the columns to merge on
df_churn_temp = df_temp.merge(df_contract, left_on="CustomerId", right_on="customerID")
df_churn_temp.info()
```

```python
# Another approach to merge DataFrames
df_churn = df_customer.merge(df_services, on="CustomerId").merge(df_contract, left_on="CustomerId", right_on="customerID")
df_churn.drop(["customerID"], axis=1, inplace=True)
df_churn.info()
```

## Taking care of missing values

What is the solution for missing values?
- 1: Remove the missing values, the value, row or column
- 2: Substitute the missing values with a constant value

```python
# Remove the 'TotalCharges' column from the DataFrame
df_churn.drop(columns=['TotalCharges'], axis=1)
```

```python
# Or if we donÂ´t know the column name
df_churn.dropna(axis=1)
```

```python
# We also can remove the column only if all the values in that column are null.
# In this example nothing happens, because only 11 lines on TotalCharges column are null.
df_churn.dropna(axis=1, how='all')
```

```python
# Remove the lines with missing values
# With that, the 11 lines with missing values are removed from the dataset.
df_churn.dropna(axis=0)
```

```python
# Remove the lines with all columns as NA values
# Nothing happens
df_churn.dropna(axis=0, how='all')
```

```python
# Let's try to fix the missing values, changing to zero
df_churn.fillna(0)
```

```python
# Sometimes we have categoric columns, and we to fill with a correct value
df_churn.fillna(value={'TotalCharges': 0, 'Contract': 'N/A'})
```

```python
# Here we can use a formula as missing value, we can use mode, median or mean to fill the missing values
mean = df_churn.TotalCharges.mean()
df_churn.fillna(value={'TotalCharges': mean})
```
